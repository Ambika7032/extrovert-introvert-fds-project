```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
df = pd.read_csv("C:/Users/ROHIT/Documents/sathvika/archive (12)/personality_dataset.csv") 
df.head()
```

```python
df.shape
df.info()
df.describe()
df['Personality'].value_counts()
```

```python
df.dropna(inplace=True)  # Drop missing values

# Convert categorical responses to numerical (if needed)
df['Personality'] = df['Personality'].astype('category')
```

```python
df['Personality'].value_counts().plot.pie(autopct='%1.1f%%', colors=sns.color_palette('pastel'))
plt.title("Extrovert vs Introvert Distribution")
plt.ylabel("")
```

```python
sns.scatterplot(x='Social_event_attendance', y='Friends_circle_size', hue='Personality', data=df)
plt.title("Event Attendance vs Friends Circle Size")
```

```python
sns.boxplot(x='Personality', y='Time_spent_Alone', data=df)
plt.title("Time Spent Alone by Personality")
```

```python
sns.histplot(data=df, x='Going_outside', hue='Personality', bins=10, kde=True, multiple='stack')
plt.title("Going Outside Frequency")
```

```python
sns.lineplot(data=df, x='Post_frequency', y='Stage_fear', hue='Personality', marker='o')
plt.title("Post Frequency vs Stage Fear by Personality")
```

```python
# Select only numeric columns
numeric_df = df.select_dtypes(include=['int64', 'float64'])

# Plot heatmap
sns.heatmap(numeric_df.corr(), annot=True, cmap='coolwarm')
plt.title("Correlation Between Numeric Features")
plt.show()
```
